geography: germany=1,spain=2,france=3
gender: male=1,female=0

```python
from tkinter import *
from sklearn.preprocessing import StandardScaler
import joblib

def show_entry_fields():
    try:
        p1 = int(e1.get())
        p2 = int(e2.get())
        p3 = int(e3.get())
        p4 = float(e4.get())
        p5 = int(e5.get())
        p6 = int(e6.get())
        p7 = int(e7.get())
        p8 = float(e8.get())
        p9 = int(e9.get())  # Geography
        p10 = int(e10.get())  # Gender

        # Geography encoding
        Geography_Germany = 1 if p9 == 1 else 0
        Geography_Spain = 1 if p9 == 2 else 0
        Geography_France = 1 if p9 == 3 else 0

        model = joblib.load('churn_prediction_model')  
        sc = joblib.load('scaler.save')              

        inputs = [[p1, p2, p3, p4, p5, p6, p7, p8, Geography_Germany, Geography_Spain, p10]]
        inputs_scaled = sc.transform(inputs)

        result = model.predict(inputs_scaled)

        if result == 0:
            prediction_label.config(text="Prediction: No Exit", fg="green")
        else:
            prediction_label.config(text="Prediction: Exit", fg="red")
    except Exception as e:
        prediction_label.config(text=f"Error: {str(e)}", fg="red")

# Main window
master = Tk()
master.title("Bank Customer Churn Prediction")
master.geometry("700x600")
master.configure(bg="#f2f2f2")

# Allow grid resizing
master.grid_rowconfigure(0, weight=1)
master.grid_columnconfigure(0, weight=1)

# Frame to hold all widgets and center them
main_frame = Frame(master, bg="#f2f2f2")
main_frame.grid(row=0, column=0, sticky="nsew", padx=20, pady=20)

# Allow frame content to expand
for i in range(13):
    main_frame.grid_rowconfigure(i, weight=1)
main_frame.grid_columnconfigure(0, weight=1)
main_frame.grid_columnconfigure(1, weight=2)

# Title
Label(main_frame, text="Customer Churn Prediction Using ML", font=("Helvetica", 16, "bold"),
      bg="#4a4a4a", fg="white", pady=10).grid(row=0, column=0, columnspan=2, pady=(10, 20), sticky="ew")

# Labels and Entries
labels = [
    "CreditScore", "Age", "Tenure", "Balance", "NumOfProducts",
    "HasCrCard", "IsActiveMember", "EstimatedSalary",
    "Geography (1=Germany, 2=Spain, 3=France)", "Gender (1=Male, 2=Female)"
]

entries = []

for i, text in enumerate(labels):
    Label(main_frame, text=text, anchor='w', bg="#f2f2f2").grid(row=i+1, column=0, sticky='w', pady=5)
    entry = Entry(main_frame)
    entry.grid(row=i+1, column=1, sticky='ew', pady=5)
    entries.append(entry)

e1, e2, e3, e4, e5, e6, e7, e8, e9, e10 = entries

# Predict button
Button(main_frame, text='Predict', command=show_entry_fields, bg="#4CAF50", fg="white",
       font=("Helvetica", 10, "bold"), padx=10, pady=5).grid(row=11, column=0, columnspan=2, pady=20)

# Prediction label
prediction_label = Label(main_frame, text="", font=("Helvetica", 12, "bold"), bg="#f2f2f2")
prediction_label.grid(row=12, column=0, columnspan=2)

master.mainloop()
```
